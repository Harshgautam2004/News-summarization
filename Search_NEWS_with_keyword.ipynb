```python
#storing all url's together
url_link=[]
for article in all_articles['articles']:
    url_link.append(article['url'])
print(url_link)
```

```python
import bs4 as bs
#urllib is used to access html page 
from urllib.request import Request, urlopen
import re
```

```python
hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) ' 
                        'AppleWebKit/537.11 (KHTML, like Gecko) '
                        'Chrome/23.0.1271.64 Safari/537.11',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
            'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
            'Accept-Encoding': 'none',
            'Accept-Language': 'en-US,en;q=0.8',
            'Connection': 'keep-alive'}
#extracting all paragraphs from the urls
article_text=""
for url in url_link:
    request = Request(url, headers=hdr) #request
    scraped_data = urlopen(request) #
    article=scraped_data.read()
    #article is simple html code
    parsed_article=bs.BeautifulSoup(article, 'lxml')
    paragraphs=parsed_article.find_all('p')
    for p in paragraphs:
        article_text+=p.text
```

```python
print(len(article_text))
article_text
```

The above article_text contains large amount of news

Now we will summarize the above news

```python
#remove numbers
article_text=re.sub(r'[[0-9]*]', ' ', article_text)
#remove whitespaces
article_text=re.sub(r'/\s+/g', ' ', article_text)
```

```python
article_text
```

```python
#removing special character and digits
formatted_article_text=re.sub('[^a-zA-Z]', ' ', article_text)
```

```python
formatted_article_text
```

```python
#Tokenising the text
import nltk
nltk.download('punkt')
sentence_list=nltk.sent_tokenize(article_text)
```

```python
sentence_list
```

```python
nltk.download('stopwords')
stopwords=nltk.corpus.stopwords.words('english')
word_frequencies={}
#remove stopwords because they are of not any use to us
for word in nltk.word_tokenize(formatted_article_text):
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word]=1
        else:
            word_frequencies[word]+=1
        
```

```python
#to check word with maximum frequency
maximum_frequency=max(word_frequencies.values())
for word in word_frequencies.keys():
    word_frequencies[word]=(word_frequencies[word]/maximum_frequency)
```

```python
word_frequencies
```

```python
#calculating scores for each sentence to prioritise them according to score
sentence_scores={}
for sent in sentence_list:
    for word in nltk.word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' '))<30:
                if sent not in sentence_scores.keys():
                    sentence_scores[sent]=word_frequencies[word]
                else:
                    sentence_scores[sent]+=word_frequencies[word]
```

```python
sentence_scores
```

```python
sizeofsen=len(sentence_scores)
print(sizeofsen)
#now taking top n/10 news
sizeofsen=int(sizeofsen/10)
```

```python
import heapq
summary_sentences=heapq.nlargest(sizeofsen, sentence_scores, key=sentence_scores.get)
summary_sentences
```

```python
summary=' '.join(summary_sentences)
```
